```python

from pyspark.sql import SparkSession

# Initialize Spark Session
spark = SparkSession.builder \
    .appName("LocalSparkSQL") \
    .config("spark.sql.shuffle.partitions", "4") \
    .getOrCreate()
```

```python
df = spark.read.csv("salesdaily.csv", header=True, inferSchema=True)
df.createOrReplaceTempView("salesdaily")
df.show()
```

```python
sales_daily = spark.sql(
    """
    select
     to_date(datum, 'M/d/yyyy') as sales_date
    ,left(to_date(datum, 'M/d/yyyy'), 7) as year_month
    ,M01AB as ab_sales
    ,M01AE as ae_sales
    ,N02BA as ba_sales
    ,N02BE as be_sales
    ,N05B as b_sales
    ,N05C as c_sales
    ,r03 as r3_sales
    ,r06 as r6_sales
    from salesdaily
    """
)
sales_daily.createOrReplaceTempView('sales_daily')
sales_daily.show()
```

```python
spark.sql(
    """
    select
    year_month
    ,sum(ab_sales)
    ,sum(ae_sales)
    ,sum(ba_sales)
    ,sum(be_sales)
    ,sum(b_sales)
    ,sum(c_sales)
    ,sum(r3_sales)
    ,sum(r6_sales)
    from
    sales_daily
    group by year_month
    order by year_month
    """
).show()
```
